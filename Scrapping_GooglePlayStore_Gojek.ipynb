```python
from google_play_scraper import app, reviews, Sort
import pandas as pd
import numpy as np
from time import sleep
```

```python
all_reviews = []
continuation_token = None

for _ in range(10):  # Loop 10x (ambil sekitar 1000 review)
    reviews_batch, continuation_token = reviews(
        'com.gojek.app',
        lang='id',
        country='id',
        sort=Sort.NEWEST,
        count=1000,  # Max 100 per request
        continuation_token=continuation_token
    )
    all_reviews.extend(reviews_batch)
    if not continuation_token:  # Stop jika tidak ada data lagi
        break
    sleep(1)  # Delay untuk hindari blokir

df_reviews = pd.DataFrame(all_reviews)
```

```python
df = df_reviews[['userName','score','at','content']]
df.head()
```
